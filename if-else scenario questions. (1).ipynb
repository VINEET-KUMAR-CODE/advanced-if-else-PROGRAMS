# 10 advanced if-else scenario questions.

## 1. Hospital Emergency Room Triage

```python
temp = float(input("Enter patient's temperature (°F): "))
pain = input("Enter pain level (none / moderate / severe): ").lower()

if temp > 102 or pain == "severe":
    priority = "High Priority"
elif (99 <= temp <= 102) or pain == "moderate":
    priority = "Medium Priority"
else:
    priority = "Low Priority"

print("Patient Priority:", priority)
```

## 2. Shopping Cart Free Delivery

```python
cart_value = float(input("Enter cart value: "))
membership = input("Do you have membership (yes/no): ").lower()

if cart_value > 2000:
    print("Delivery: Free")
elif 1000 <= cart_value <= 2000 and membership == "yes":
    print("Delivery: Free")
else:
    print("Delivery Fee: ₹100")
```

# 3. Library Fine System

```python
days = int(input("Enter number of overdue days: "))

if days <= 5:
    fine = days * 2
elif days <= 10:
    fine = days * 5
else:
    fine = days * 10

print("Total Fine: ₹", fine)
```

## 4. Airport Luggage Check

```python
travel_class = input("Enter class (economy/business/first): ").lower()
weight = float(input("Enter luggage weight (kg): "))

if travel_class == "economy":
    limit = 20
    fee_per_kg = 200
elif travel_class == "business":
    limit = 30
    fee_per_kg = 150
else:
    limit = 40
    fee_per_kg = 100

if weight > limit:
    extra = weight - limit
    fee = extra * fee_per_kg
else:
    fee = 0

print("Excess Luggage Fee: ₹", fee)
```

## 5. Hotel Room Pricing

```python
customer = input("Customer type (regular/member/VIP): ").lower()
room = input("Room type (standard/deluxe): ").lower()
price = float(input("Enter room price: "))

if customer == "regular":
    discount = 0.10
elif customer == "member":
    if room == "deluxe":
        discount = 0.20
    else:
        discount = 0.15
else:  # VIP
    discount = 0.30

final_price = price - (price * discount)
print("Final Room Price: ₹", final_price)
```

##  6. Energy Consumption Billing

```python
units = int(input("Enter total units consumed: "))

if units <= 100:
    bill = units * 5

elif units <= 300:
    bill = (100 * 5) + ((units - 100) * 10)

else:
    bill = (100 * 5) + (200 * 10) + ((units - 300) * 15)

print("Total Electricity Bill: ₹", bill)
```

## 7. Ride Fare Calculation

```python
time = input("Time of ride (day/night): ").lower()
distance = float(input("Enter distance (km): "))

if time == "day":
    rate = 15
else:
    rate = 20

fare = distance * rate

if distance > 20:
    fare *= 0.90   # 10% discount

print("Total Ride Fare: ₹", fare)
```

## 8. Event Entry Rules

```python
age = int(input("Enter age: "))
student_id = input("Student ID (yes/no): ").lower()
price = 100  # assume full price = 100

if age < 12 or age > 60:
    entry_fee = 0
elif 12 <= age <= 25 and student_id == "yes":
    entry_fee = price * 0.50
else:
    entry_fee = price

print("Entry Fee: ₹", entry_fee)
```

## 9. Parking Lot Charges

```python
vehicle = input("Vehicle type (car/bike): ").lower()
hours = int(input("Enter parking hours: "))

if vehicle == "car":
    first_hour = 50
    extra_rate = 20
else:
    first_hour = 20
    extra_rate = 10

if hours <= 1:
    fee = first_hour
else:
    fee = first_hour + (hours - 1) * extra_rate

print("Total Parking Fee: ₹", fee)
```

# 10. Real Estate Loan Eligibility

```python
income = int(input("Enter monthly income: "))
loan_status = input("Any existing loan (yes/no): ").lower()

if income > 100000 and loan_status == "no":
    ltv = "90% LTV Eligible"
elif 50000 <= income <= 100000 or loan_status == "yes":
    ltv = "70% LTV Eligible"
else:
    ltv = "50% LTV Eligible"

print(ltv)
```
